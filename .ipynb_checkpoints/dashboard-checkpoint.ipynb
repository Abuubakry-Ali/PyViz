# Toronto Dwellings Analysis Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions to create a Panel dashboard.

```python
# initial imports
import os
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
import plotly.express as px
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Initialize the Panel Extensions (for Plotly)
import panel as pn
pn.extension("plotly")
```

```python
# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("mapbox")
px.set_mapbox_access_token(map_box_api)
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/toronto_neighbourhoods_census_data.csv")
to_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/toronto_neighbourhoods_coordinates.csv")
df_neighbourhood_locations = pd.read_csv(file_path)
```

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

### Global available data

```python
# Getting the data from the top 10 expensive neighbourhoods
# YOUR CODE HERE!
top_10_expensive_neig=to_data.groupby(['neighbourhood']).mean()
top_10_expensive_neig=top_10_expensive_neig.sort_values(
    ['average_house_value'],
    ascending=False
     ).reset_index()[:10]

# Calculate the mean number of dwelling types units per year
# YOUR CODE HERE!
dwelling_types = to_data.drop(columns=['average_house_value','shelter_costs_rented','shelter_costs_owned'])
dwelling_types = dwelling_types.reset_index()
dwelling_type=dwelling_types.groupby('year').sum()

# Calculate the average monthly shelter costs for owned and rented dwellings
# YOUR CODE HERE!
average_monthly_shelter=to_data.groupby(['year'])[['shelter_costs_owned','shelter_costs_rented']].mean()
```

### Panel Visualization Functions

```python
# Define Panel visualization functions
def neighbourhood_map():
    """Neighbourhood Map"""
    
    # YOUR CODE HERE!
    grouped_neighborhood = to_data.groupby('neighbourhood').mean()
    all_neighborhoods=grouped_neighborhood.reset_index()
    all_neighborhoods = pd.merge(df_neighbourhood_locations, all_neighborhoods, on='neighbourhood')
    
    px.set_mapbox_access_token(map_box_api)
    map_scatter = px.scatter_mapbox(all_neighborhoods,
    lat='lat',
    lon='lon',
    size='average_house_value',
    color='average_house_value',
    color_continuous_scale=px.colors.cyclical.IceFire,
    size_max=15,
    zoom=11,
    hover_name='neighbourhood',
    title='average house price in Toronto')
    return map_scatter

def create_bar_chart(data, title, xlabel, ylabel, color):
    """
    Create a barplot based in the data argument.
    """
    
    # YOUR CODE HERE!
    data.plot(kind="bar", title=title,color=color)
    plt.xlabel(xlabel, fontsize=15)
    plt.ylabel(ylabel)
    return plt.show()
    
def create_line_chart(data, title, xlabel, ylabel, color):
    """
    Create a line chart based in the data argument.
    """
    
    # YOUR CODE HERE!
    data.plot(kind="line",color=color)
    plt.title(title,fontsize=15, weight='bold')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    return plt.show()
    
def average_house_value():
    """Average house values per year."""

    # YOUR CODE HERE!
    ave_house_value=to_data.groupby(['year'])['average_house_value'].mean()
    return create_line_chart(ave_house_value,title='average_house_value', xlabel='year', ylabel='Cost',color='blue')
    
def average_value_by_neighbourhood():
    """Average house values by neighbourhood."""
    
    # YOUR CODE HERE!
    house_values_neighbour =to_data[['neighbourhood','average_house_value']].reset_index()
    return  house_values_neighbour.hvplot(x='year', y='average_house_value', groupby='neighbourhood', kind='line')
    
    
def number_dwelling_types():
    """Number of dwelling types per year"""
    
    # YOUR CODE HERE!
    column=['single_detached_house','apartment_five_storeys_plus','movable_dwelling','semi_detached_house','row_house','duplex','apartment_five_storeys_less','other_house']
    dwelling_type=dwelling_types.hvplot(
    x='year',
    y=column,
    rot = 90,
    groupby='neighbourhood',
    kind='bar',
    width=900,
    height=500)
    
    return dwelling_type

def average_house_value_snapshot():
    """Average house value for all Toronto's neighbourhoods per year."""

    # YOUR CODE HERE!
   
    house_values_toronto=to_data.groupby(
    ['year',"neighbourhood",'average_house_value']).mean().drop(columns=['shelter_costs_rented','shelter_costs_owned'])
    plot = px.bar(
    house_values_toronto.reset_index(),
    x="neighbourhood",
    y="average_house_value",
    color="average_house_value",
    title="",
    facet_row="year")
    return plot


def top_most_expensive_neighbourhoods():
    """Top 10 most expensive neighbourhoods."""
    
    # YOUR CODE HERE!
    top_10_expensive= top_10_expensive_neig.hvplot(
    x='neighbourhood',
    y='average_house_value', 
    rot = 90,
    kind='bar',
    title='Top 10 expensive neighbourhoods in Toronto',
    width=1000,
    height=600)
    return top_10_expensive

def sunburts_cost_analysis():
    """Sunburst chart to conduct a costs analysis of most expensive neighbourhoods in Toronto per year."""
    
    # YOUR CODE HERE!
    df_top_10_2001 = to_data.loc[2001].sort_values(['average_house_value'],ascending=False)[:10].reset_index()
    df_top_10_2006 =to_data.loc[2006].sort_values(['average_house_value'],ascending=False)[:10].reset_index()
    df_top_10_2011 =to_data.loc[2011].sort_values(['average_house_value'],ascending=False)[:10].reset_index()
    df_top_10_2016 =to_data.loc[2016].sort_values(['average_house_value'],ascending=False)[:10].reset_index()

    top_10_all_expensive = pd.concat([df_top_10_2001,df_top_10_2006,df_top_10_2011,df_top_10_2016],axis=0,join='inner')
    
    fig = px.sunburst(
    top_10_all_expensive,
    path=['year', 'neighbourhood'], 
    color='shelter_costs_owned',
    values='average_house_value',
    title='Costs analysis of most expensive neighbourhoods in Toronto per year')
    return fig.show()
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
# YOUR CODE HERE!

title = '#Real Estate Analysis of Toronto from 2001-2016'

# Define a welcome text
# YOUR CODE HERE!

welcome_tab = pn.Column ('This Dashboard present Real Estate Analysis of Toronto from 2001-2016 ', neighbourhood_map())


# Create a tab layout for the dashboard
# YOUR CODE HERE!
                     
most_expensive_neighbourhoods = pn.Row(
    "## Top 10 most expensive neighbourhood ",top_most_expensive_neighbourhoods(),sunburts_cost_analysis())
                     
yearly_market= pn.Row(
    create_bar_chart(dwelling_type.loc[2001], "Dwelling Types in Toronto in 2001", "2001", "Dwelling Type Units", "red"),
    create_bar_chart(dwelling_type.loc[2006], "Dwelling Types in Toronto in 2006", "2006", "Dwelling Type Units", "blue"),
    pn.Row(create_bar_chart(dwelling_type.loc[2011], "Dwelling Types in Toronto in 2011", "2011", "Dwelling Type Units", "orange"),
    create_bar_chart(dwelling_type.loc[2016], "Dwelling Types in Toronto in 2016", "2016", "Dwelling Type Units", "magenta"),
))
neighbourhood_analysis = pn.Row(
    pn.Column(average_value_by_neighbourhood(),
    number_dwelling_types()),
    average_house_value_snapshot()
)

shelter_cost_house_value = pn.Column(
    create_line_chart(average_monthly_shelter["shelter_costs_rented"], "Average Monthly Shelter Cost for Rented Dwellings in Toronto", "Year", "Avg Monthly Shelter Costs", "orange"),
    create_line_chart(average_monthly_shelter["shelter_costs_owned"], "Average Monthly Shelter Cost for Owned Dwellings in Toronto", "Year", "Avg Monthly Shelter Costs", "blue"),
    average_house_value(),
)

# Create the main dashboard
# YOUR CODE HERE!

dashboard = pn.Column( title,pn.Tabs(
    ("Welcome",welcome_tab),("Yearly Markrt Analysis",yearly_market),
   ('Shelter cost vs house value ',shelter_cost_house_value), 
    ("Neighbourhood Analysis", neighbourhood_analysis), 
    ("Top expensive neighbourhoods", most_expensive_neighbourhoods)
))
```

```python
## Serve the Panel Dashboard
dashboard.servable()
```
